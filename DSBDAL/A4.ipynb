```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np  
# df=read_csv('') 
df=sns.load_dataset("titanic")  
df
```

```python
df.info() 
df.isna().sum() 
```

```python
# Drop rows with null values in the "Age" column
df = df.dropna(subset=['age'])
```

```python
df.info()
```

```python
df=df.astype({
    'age':'int64'
})
```

```python
df.dtypes
```

```python
# A count plot of Passenger Class in titanic  
sns.countplot(data=df, x='pclass' )
```

```python

sns.countplot(data=df, x='pclass', hue='survived')

plt.xlabel('Passenger Class')

plt.legend(title='age')

plt.show()
```

```python
# A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. 
# KDE represents the data using a continuous probability density curve in one or more dimensions.  
sns.histplot(data=df, x='age', bins=10, kde=True, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Passenger Ages')
plt.show()
```
