```python
# Outlier Detection -> IQR
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

df = df[(df['math score'] > lower) & (df['math score'] < upper)]
df.shape
```

```python
sns.boxplot(df)
```

```python
# Removing the outliers of the writing score  
#let's check the shape  
df.shape
```

```python
Q1=df['writing score'].quantile(0.25) 
Q3=df['writing score'].quantile(0.75) 
IQR=Q3-Q1 
lower=Q1-1.5*IQR      
upper=Q3+1.5*IQR  
new_df=df[(df['writing score']>lower)&(df['writing score']<upper)]  
new_df.shape
```

```python
sns.boxplot(new_df['writing score'])
```
